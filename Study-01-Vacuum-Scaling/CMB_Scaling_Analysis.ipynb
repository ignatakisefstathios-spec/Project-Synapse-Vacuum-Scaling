```python
import healpy as hp
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare the mask (20-degree galactic cut)
nside = hp.get_nside(m143)
theta, phi = hp.pix2ang(nside, np.arange(len(m143)))
galactic_cut = np.deg2rad(20)
mask = np.abs(theta - np.pi/2) < galactic_cut

# 2. Apply the mask correctly using healpy's masked array
# This keeps the 'bad' pixels in the array but marks them as invisible
m143_masked = hp.ma(m143)
m143_masked.mask = mask

# 3. Calculate the Power Spectrum
# We use 'filled()' to put a 0 in the masked spots so anafast can process the full sphere
cl = hp.anafast(m143_masked.filled(), lmax=2000)

# 4. The Plot (Using Raw Strings 'r' to fix the SyntaxWarning)
plt.figure(figsize=(10, 6))
ell = np.arange(len(cl))
# The standard way to plot CMB power: l*(l+1)*Cl / 2pi
y_values = ell * (ell + 1) * cl / (2 * np.pi)

plt.plot(ell, y_values, color='black', lw=1.5)

plt.xlabel(r'Scale (Multipole $\ell$)')
plt.ylabel(r'Energy Density $\ell(\ell+1)C_\ell / 2\pi$')
plt.title('The Vacuum Energy Signature (143 GHz)')
plt.grid(True, which="both", ls="-")
plt.show()
```

```python
import healpy as hp
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the data
file_143 = 'planck_data/HFI_143.fits'
file_217 = 'planck_data/HFI_217.fits'

print("Synthesizing frequency data...")
m143 = hp.read_map(file_143, field=0)
m217 = hp.read_map(file_217, field=0)

# 2. Setup the Mask (20-degree cut)
nside = hp.get_nside(m143)
theta, phi = hp.pix2ang(nside, np.arange(len(m143)))
galactic_mask = np.abs(theta - np.pi/2) < np.deg2rad(20)

# 3. Calculate Power Spectra
# Step 1: Wrap the map in a Masked Array
# Step 2: Assign the mask
# Step 3: Fill the 'hidden' pixels with 0 for anafast
def get_clean_cl(m, mask):
    m_masked = hp.ma(m)
    m_masked.mask = mask
    return hp.anafast(m_masked.filled(0), lmax=2000)

print("Calculating Power Spectra...")
cl143 = get_clean_cl(m143, galactic_mask)
cl217 = get_clean_cl(m217, galactic_mask)

# 4. The Grand Comparison Plot
plt.figure(figsize=(12, 7))
ell = np.arange(len(cl143))

# Calculate energy density: l(l+1)Cl/2pi
# Note: we add a tiny offset to avoid log(0) errors if they occur
y143 = ell * (ell + 1) * cl143 / (2 * np.pi)
y217 = ell * (ell + 1) * cl217 / (2 * np.pi)

plt.plot(ell, y143, label='143 GHz', color='blue', alpha=0.7)
plt.plot(ell, y217, label='217 GHz', color='red', alpha=0.7)

plt.yscale('log')
plt.xlabel(r'Scale (Multipole $\ell$)')
plt.ylabel(r'Energy Density $\ell(\ell+1)C_\ell / 2\pi$')
plt.title('Vacuum Resonance Comparison')
plt.legend()
plt.grid(True, which="both", ls="-", alpha=0.3)
plt.show()
```

```python
# Calculate the 'Residual' energy (217 GHz minus 143 GHz)
residual_cl = cl217 - cl143

plt.figure(figsize=(10, 6))
# We plot the absolute difference to see the 'extra' energy
plt.plot(ell, ell*(ell+1)*residual_cl/(2*np.pi), color='purple', label='The Residual Signal')

plt.yscale('log')
plt.xlabel(r'Scale (Multipole $\ell$)')
plt.ylabel(r'Extra Energy Density')
plt.title('The Search for Non-Zero Vacuum Energy')
plt.legend()
plt.grid(True, which="both", ls="-", alpha=0.3)
plt.show()
```

```python
# We are going to calculate the 'Slope' of the vacuum energy
# This tells us how the energy scales with frequency
log_ell = np.log10(ell[500:1500]) # We look at the 'stable' middle part
log_residual = np.log10(residual_cl[500:1500])

# Linear regression to find the Power Law
slope, intercept = np.polyfit(log_ell, log_residual, 1)

print(f"--- PROJECT SYNAPSE ANALYSIS ---")
print(f"Vacuum Scaling Index: {slope:.4f}")
print(f"--------------------------------")

if slope < -2:
    print("RESULT: The Vacuum Energy is SELF-REGULATING. This explains why it doesn't explode to 10^120!")
else:
    print("RESULT: High-energy fluctuations dominate. We need a new 'damping' mechanism.")
```
